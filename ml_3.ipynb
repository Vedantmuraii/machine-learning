```python
dataframe.drop(['PassengerId','SibSp','Name','Embarked','Parch'], inplace=True, axis=1)
```

```python
dataframe.info()
```

```python
dataframe.drop(['Cabin',], inplace=True, axis=1)
```

```python
data=dataframe.drop('Survived',axis='columns')
```

```python
target=dataframe.Survived
```

```python
print(target)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.Age=data.Age.fillna(data.Age.mean())
```

```python
data.Sex=data.Sex.replace("male",0)
data.Sex=data.Sex.replace("female",1)
```

```python
data.info()
```

```python
data.head()
```

```python
data.drop(['Ticket',], inplace=True, axis=1)
```

```python
target.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data,target,test_size=0.2)
# x dependent , y indepenndent
```

```python
len(x_train)
```

```python
x_train.shape
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict([[1,2,38.0,70.23]])
```

```python
model.score(x_test,y_test)
```
